```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('customers.csv')
df = df.drop(['Channel','Region'],axis=1)
df.dropna()
df
```

```python
data = np.array(df)
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(40, 60))
plt.title("Customer Dendograms - Single Linkage")  
dend = shc.dendrogram(shc.linkage(data, method='single'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=8, affinity='euclidean', linkage='single')  
single = cluster.fit_predict(data)
```

```python
plt.figure(figsize=(40, 60))  
plt.title("Customer Dendograms - Average Linkage")  
dend = shc.dendrogram(shc.linkage(data, method='average'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='average')  
average = cluster.fit_predict(data)
```

```python
plt.figure(figsize=(40, 60))  
plt.title("Customer Dendograms - Complete Linkage")  
dend = shc.dendrogram(shc.linkage(data, method='complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')  
complete = cluster.fit_predict(data)
```

```python
plt.figure(figsize=(40, 60))  
plt.title("Customer Dendograms - Ward Linkage")  
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='ward')  
ward = cluster.fit_predict(data)
ward
```
